```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('/content/test.csv')
dataset.head()
```

```python
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in features_nan:
  
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,features_nan)

dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
for feature in numerical_with_nan:
  print('{}: {}% Missing value'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in numerical_with_nan:
  median_value=dataset[feature].median()
  dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
  dataset[feature].fillna(median_value,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
import numpy as np
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for feature in num_features:
  dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
categorical_feature =[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
categorical_feature
```

```python
for feature in categorical_feature:
  temp = dataset.groupby(feature)['Id'].count()/len(dataset)
  temp_df=temp[temp>0.01].index
  dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_value')
```

```python
dataset.head(50)
```

```python
for feature in categorical_feature:
  labels_orederd=dataset.groupby([feature])['Id'].mean().sort_values().index
  labels_orederd={k: i for  i,k in enumerate(labels_orederd,0)}
  dataset[feature]=dataset[feature].map(labels_orederd)
```

```python
dataset.head(10)
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
len(scaling_feature)
```

```python
scaling_feature
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[scaling_feature])
```

```python
scaler.transform(dataset[scaling_feature])
```

```python
data=pd.concat([dataset[['Id']].reset_index(drop=True),
               pd.DataFrame(scaler.transform(dataset[scaling_feature]),columns=scaling_feature)],
               axis=1)
```
